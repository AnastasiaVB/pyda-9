Задание 1

Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
# ratings.head()
movies = pd.read_csv('movies.csv')
# movies.head()
```

```python
# ratings.info()
ratings_movies = ratings.merge(movies, on = 'movieId', how='left')
# ratings_movies.head()

rating5_movies = ratings_movies[ratings_movies['rating']==5]
# rating5_movies.head()

rating5_movies['title'].value_counts().head(1)
```

Задание 2

По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
data = pd.read_csv('power.csv')
```

```python
# data.head()

data_filtered = data[data['country'].isin(['Lithuania','Estonia','Latvia']) & data['category'].isin([4,12,21]) & (data['year']<2010) & (data['year']>2005) & data['quantity']>0]

data_filtered.quantity.sum()

```
